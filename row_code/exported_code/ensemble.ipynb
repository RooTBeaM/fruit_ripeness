```python
import pandas as pd

pred1 = pd.read_csv('test_sim_vit_512_384_tv_noflip.csv')
pred1_score = pd.read_csv('test_sim_vit_512_384_tv_noflip_score.csv')
pred2 = pd.read_csv('test_sim_vit_512_384_tv_flip.csv')
pred2_score = pd.read_csv('test_sim_vit_512_384_tv_flip_score.csv')
```

```python
pred1 = pred1[['path', '1']]
pred1.columns = ['path', 'p1']
pred1['score1'] = pred1_score['1']
```

```python
pred2 = pred2[['path', '1']]
pred2.columns = ['path', 'p2']
pred2['score2'] = pred2_score['1']
```

```python
endf = pd.concat([pred1, pred2[['p2', 'score2']]], axis=1)[['path', 'p1', 'p2', 'score1', 'score2']]
```

```python
endf[(endf['score1'] > endf['score2']) & (endf['p1'] != endf['p2'])]
```

```python
endf[(endf['score1'] < endf['score2']) & (endf['p1'] != endf['p2'])]
```
